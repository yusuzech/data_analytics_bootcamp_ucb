# Scrape Craglist

## Importing Modules

```python
#to install
#!pip install beautifulsoup4
!pip insatll requests
```

```python
# Dependencies
from bs4 import BeautifulSoup
import requests
```

## Making Requests

```python
# URL of page to be scraped
url = 'https://newjersey.craigslist.org/search/sss?sort=rel&query=guitar'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

## Parsing Results

```python
# results are returned as an iterable list
results = soup.find_all('li', class_="result-row")
```

```python
# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        title = result.find('a', class_="result-title").text
        # Identify and return price of listing
        price = result.a.span.text
        # Identify and return link to listing
        link = result.a['href']

        # Print results only if title, price, and link are available
        if (title and price and link):
            print('-------------')
            print(title)
            print(price)
            print(link)
    except AttributeError as e:
        print(e)
```

# Export Web-Scraping Result to MongoDB

## Importing Modules

```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pymongo
```

## Set up MongoDB

Before you start, please open your terminal and run `mongod` to start the server.

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.craigslist_db
collection = db.items
```

## Making Requests

```python
# URL of page to be scraped
url = 'https://newjersey.craigslist.org/search/sss?sort=rel&query=guitar'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')
```

## Parsing and Saving results to MongoDB

```python
# Examine the results, then determine element that contains sought info
# results are returned as an iterable list
results = soup.find_all('li', class_='result-row')

# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        title = result.find('a', class_='result-title').text
        # Identify and return price of listing
        price = result.a.span.text
        # Identify and return link to listing
        link = result.a['href']

        # Run only if title, price, and link are available
        if (title and price and link):
            # Print results
            print('-------------')
            print(title)
            print(price)
            print(link)

            # Dictionary to be inserted as a MongoDB document
            post = {
                'title': title,
                'price': price,
                'url': link
            }

            collection.insert_one(post)

    except Exception as e:
        print(e)
```

## Inspecting Results

```python
# Display items in MongoDB collection
listings = db.items.find()

for listing in listings:
    print(listing)
```

## Reading Tables

We can use the `read_html` function in Pandas to automatically scrape any tabular data from a page.

```python
url = 'https://en.wikipedia.org/wiki/List_of_capitals_in_the_United_States'
```

```python
tables = pd.read_html(url)
```

```python
# seven tables in this page
print("Number of tables in this page:" + str(len(tables)))
print("Type of variable 'tables': "+ str(type(tables)))
```

```python
for table in tables:
    display(table.head())
```

We can slice off any of those dataframes that we want using normal indexing.

```python
df = tables[0]
df.columns = ['State', 'Abr.', 'State-hood Rank', 'Capital', 
              'Capital Since', 'Area (sq-mi)', 'Municipal Population', 'Metropolitan', 
              'Metropolitan Population', 'Population Rank', 'Notes']
df.head()
```

Slide Type
Cleanup of extra rows

```python
df = df.iloc[2:]
df.head()
```

Set the index to the State column

```python
df.set_index('State', inplace=True)
df.head()
```

```python
df.loc['Alabama']
```

## DataFrames as HTML

Pandas also had a to_html method that we can use to generate HTML tables from DataFrames.

```python
html_table = df.to_html()
html_table
```

You may have to strip unwanted newlines to clean up the table.

```python
html_table.replace('\n', '')
```
